```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
```

```python
df = pd.read_csv('Recidivism.csv')


# Convert Recidivism into a Numeric Value
df['Recidivism_Within_3years'] = df['Recidivism_Within_3years'].map({True: 1, False: 0})

# Convert THC Positive into a Bool
df['DrugTests_THC_Positive'] = df['DrugTests_THC_Positive'].apply(lambda x: x >= 0.5)


# print(df['Recidivism_Within_3years'].unique())
smf.logit(formula="""Recidivism_Within_3years ~ Gender + Race + Prison_Offense + Education_Level + Dependents
          + Violations_ElectronicMonitoring + Violations_Instruction + Violations_FailToReport + Violations_MoveWithoutPermission
          + Delinquency_Reports + Avg_Days_per_DrugTest + DrugTests_THC_Positive + Percent_Days_Employed
          """, data=df).fit().summary()

# ISSUE WITH PANDAS OBJECT TYPE
# endog = df['Recidivism_Within_3years']
# exog = df.drop(columns=['Recidivism_Within_3years'])
# sm.Logit(endog, exog).fit().summary()
```
